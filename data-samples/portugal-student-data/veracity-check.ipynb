```python
import json
json_data = json.loads(df.to_json(orient='records'))
json_data[:5]
```

```python
import json

# Print the body JSON object
# Read the veracity-config.json file
with open('veracity-config.json') as file:
    config = json.load(file)

# Create the body JSON object
body_json = {
    'data': json_data,
    'config': config,
}
```

```python
import requests

url = 'http://localhost:3000/checks'
headers = {'Content-Type': 'application/json'}
response = requests.post(url, json=body_json, headers=headers)

print(response.status_code)
print(response.text)  # print raw response text
```

```python
import json
url = 'http://localhost:3000/checks/:id'

if response.status_code == 200:
    process_id = json.loads(response.text)['processId']
    print(process_id)
    url_with_id = url.replace(':id', process_id)
    response_with_id = requests.get(url_with_id)
    print(response_with_id.status_code)
    pretty_response = json.dumps(json.loads(response_with_id.text), indent=4)
    print(pretty_response)
else:
    print("Error: Status code is not 200")
```
