```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('student-mat.csv', sep=';')

# Display the first few rows of the DataFrame
df.head()
```

```python
import json
json_data = json.loads(df.to_json(orient='records'))
json_data
```

```python
import requests

url = 'http://localhost:3000/veracity/process'
headers = {'Content-Type': 'application/json'}
response = requests.post(url, json=json_data, headers=headers)

print(response.status_code)
print(response.text)  # print raw response text
```

```python
import json
url = 'http://localhost:3000/veracity/process/:id'

if response.status_code == 200:
    process_id = json.loads(response.text)['processId']
    print(process_id)
    url_with_id = url.replace(':id', process_id)
    response_with_id = requests.get(url_with_id)
    print(response_with_id.status_code)
    print(response_with_id.text)
else:
    print("Error: Status code is not 200")
```
